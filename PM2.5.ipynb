```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from sorted_months_weekdays import *

from sort_dataframeby_monthorweek import *
```

```python
df = pd.read_csv("./data/new_combined_pm2.csv")
```

```python
df.head()
print(df.shape)
```

```python

print(df['Daily Mean PM2.5 Concentration'].skew())
df['Daily Mean PM2.5 Concentration'].describe()
```

```python
lower_quantile, upper_quantile = df['Daily Mean PM2.5 Concentration'].quantile([.1, .9])
```

```python
df1=df.loc[(df['Daily Mean PM2.5 Concentration'] > lower_quantile) & (df['Daily Mean PM2.5 Concentration'] < upper_quantile)]
```

```python
print(df1['Daily Mean PM2.5 Concentration'].skew())
df1['Daily Mean PM2.5 Concentration'].describe()
print(df1.shape)
```

```python
df1.head()
```

```python
county_wise ={}
for c in df1['County'].unique():
    temp = df1[df1['County'] == c]
    
    county_wise[c] = temp
    fig = go.Figure([go.Scatter(x=temp['Date'], y=temp['Daily Mean PM2.5 Concentration'])])
    fig.update_layout(
    title="Concentration of PM2.5 in " +c ,
    xaxis_title="Date",
    yaxis_title="Concentration of PM2.5",
    legend_title= c,
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"))
    fig.show()

# df1['County'].unique()
```

```python
for d in county_wise:
    county_wise[d]['Date'] = pd.to_datetime(county_wise[d]['Date'], errors='coerce')
    print("------------------------")
    print("For "+ d +" AVG by month")
    s= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['Daily Mean PM2.5 Concentration'].sum()
    case= county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['New Positives'].sum()
#     print(s)
    count = county_wise[d].groupby(county_wise[d]['Date'].dt.strftime('%B'))['Daily Mean PM2.5 Concentration'].count()
#     print(count)

#     print(s/count)
    avg = pd.DataFrame(s/count)
    avg.reset_index(level=0, inplace=True)
    
    avg.columns = ["Month", "PM2.5 Concentration"]
    avg["Total PM2.5"] =s.values 
    avg["count"] = count.values
    avg["Total Cases"] = case.values
    avg["Avg Cases"] = (case/count).values
    
    avg =Sort_Dataframeby_Month(df=avg,monthcolumnname='Month')
    print(avg)
    print("------------------------")
```
